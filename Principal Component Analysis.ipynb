```python
df = pd.DataFrame(breast_cancer['data'],columns = breast_cancer['feature_names'])
```

```python
df.describe()
```

```python
x = df
y = breast_cancer.target
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(x)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2) 
## if a data has multiple dimensions it has problem of underfitting)
```

```python
x_pca = pca.fit_transform(x)
```

```python
x_pca.shape
```

```python
x.shape
```

```python
x_pca
```

```python
pca.explained_variance_ratio_  ## features capturing ratio of variance of each component
```

```python
pca.n_components_
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train_pca, x_test_pca, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train_pca,y_train)
model.score(x_test_pca, y_test)
```
